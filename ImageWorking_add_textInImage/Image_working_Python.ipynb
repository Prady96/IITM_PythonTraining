```python
from PIL import Image,ImageDraw,ImageFont
```

```python
image = Image.open('testImage01.png')
# image.show()
image
```

```python
## defining the font
font_type = ImageFont.truetype('Arial.ttf',58)
font_type_2 = ImageFont.truetype('fonts_Dir/OpenSans-Bold.ttf',90)
font_type_3 = ImageFont.truetype('fonts_Dir/OpenSans-Light.ttf',158)

## draw object so that we can draw on top of our image
draw = ImageDraw.Draw(image)
## draw text on the image
draw.text(xy=(50,50),text="Text One",fill=(255,69,0),font=font_type)
draw.text(xy=(150,150),text="Text Two",fill=(255,69,0),font=font_type_2)
draw.text(xy=(250,250),text="Text Three",fill=(255,69,0),font=font_type_3)
## show the image
# image.show()
image
```
