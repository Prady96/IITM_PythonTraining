# Lists

```python
ages = [22,33,60,99] #mutable #can Chnage
print(ages)
print("Append element")
ages.append(100) #entry at the last position
print(ages)
```

```python
print("insert at specific place")
# Added Element 55 at the 1st Position 
ages.insert(1,55) #In this we inserted at a position
print(ages)
```

```python
print("Words Split")
str = "Welcome*to*python*world"
words = str.split('*')
print(words)
print(words[0:2])
```

```python
str = "Welcome in my Country"
words = str.split('')
print(words)
print(words[0:2])

# Insert words
words.insert(2,'hello')
print(words)

# Append Words
words.append("bye")
print(words)
```

```python
person = dict(  Name="Ajay",
                Age="23",
                Salary="80000") #mutable
print(person)
```

```python
print("\n") # New Line Char
# print("\t") #  Tabs space
print("Age of the Person is "+person["Age"])
print("Name of the Person is "+person["Name"])
```

```python
person["Name"]="Sharaya Gupta" #Change the Name of the Person
print(person["Name"])
print("\n")
print(person)
```

```python
print("Add Insurance Key")
person["Insurance"] = "Yes" #Added insurance key
print(person)
```

```python
# Delete Age
print("Delete Age")
del person["Age"]
print(person)
```
